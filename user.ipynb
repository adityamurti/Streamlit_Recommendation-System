```python
df_movies["popularity_prob"] = df_movies["vote_count"] / df_movies["vote_count"].sum()
```

```python
movie_ids = df_movies["id"].values
vote_avg = df_movies["vote_average"].values
pop_probs = df_movies["popularity_prob"].values
```

```python
np.random.seed(42)

sampled_indices = np.random.choice(
    len(df_movies),
    size=len(df_amazon_ratings),
    replace=True,
    p=pop_probs
)

df_amazon_ratings["id"] = df_movies.iloc[sampled_indices]["id"].values
df_amazon_ratings["vote_average"] = df_movies.iloc[sampled_indices]["vote_average"].values
```

```python
df_amazon_ratings["rating_scaled"] = df_amazon_ratings["rating"] * 2
```

```python
df_amazon_ratings["final_rating"] = (
    df_amazon_ratings["rating_scaled"] * 0.7 +
    df_amazon_ratings["vote_average"] * 0.3
)
```

```python
df_amazon_ratings["final_rating"] = df_amazon_ratings["final_rating"].clip(0, 10)
```

```python
df_amazon_ratings["final_rating"] = df_amazon_ratings["final_rating"].round().astype(int)
```

```python
df_amazon_ratings.head()
```

```python
df_ratings = df_amazon_ratings[["user", "id", "final_rating"]]
df_ratings.head()
```

```python
df_ratings.to_csv("user_movie_ratings_transformed.csv", index=False)
```
