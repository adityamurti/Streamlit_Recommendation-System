```python
import ast

def safe_parse(x):
    if isinstance(x, list):
        return x
    if isinstance(x, dict):
        return [x]
    if isinstance(x, float) or x is None:
        return []
    if isinstance(x, str):
        x = x.strip()
        if x == "" or x == "[]" or x.lower() == "nan":
            return []
        try:
            return ast.literal_eval(x)
        except:
            return []
    return []
```

```python
df_rec['genres'] = df_rec['genres'].apply(safe_parse)
df_rec['keywords'] = df_rec['keywords'].apply(safe_parse)
```

```python
target = []

filtered_df = df_rec[df_rec['keywords'].apply(lambda x: x == target) |
              df_rec['genres'].apply(lambda x: x == target)]

filtered_df
```

```python
df_rec['genres'] = df_rec['genres'].apply(
    lambda x: [genre['name'] for genre in x]
)

df_rec['keywords'] = df_rec['keywords'].apply(
    lambda x: [keyword['name'] for keyword in x]
)

df_rec.head()
```

```python
df_rec['genres'] = df_rec['genres'].apply(lambda x: x if x != [] else '-')
df_rec['keywords'] = df_rec['keywords'].apply(lambda x: x if x != [] else '-')
```
